```haskell
import Data.ByteString.Base64
```

```haskell
import qualified Data.Attoparsec.ByteString.Char8 as Atto
```

```haskell
base64 :: Atto.Parser B.ByteString
base64 = Atto.takeWhile1 (\c -> Atto.isAlpha_ascii c || Atto.isDigit c || Atto.inClass "+/=" c)
```

```haskell
parseEntry :: Atto.Parser (Token, B.ByteString)
parseEntry = do
    bs <- base64
    Atto.space
    t <- Atto.decimal
    return (t, bs)
```

```haskell
parseFile :: Atto.Parser [(Token, B.ByteString)]
parseFile = Atto.many1 $ parseEntry <* Atto.endOfLine
```

```haskell
Atto.parse parseFile tik
```
